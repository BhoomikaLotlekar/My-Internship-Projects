```python
import pandas as pd

#loading dataset
matches = pd.read_csv("/content/IPL_matches.csv")
deliveries = pd.read_csv("/content/IPL_deliveries.csv")

matches.head()
deliveries.head()
```

displaying basic info

```python
print(matches.info())
print(deliveries.info())
```

Missing values

```python
print(matches.isnull().sum())
print(deliveries.isnull().sum())
```

IPL Team Performance Analysis

Total Wins Per Team

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x=matches['winner'].value_counts().index, y=matches['winner'].value_counts().values)
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Total Wins")
plt.title("Total Wins by Each Team in IPL")
plt.show()
```

Toss Decision Impact

```python
sns.countplot(data=matches, x="toss_won", palette="coolwarm")
plt.title("Toss Decisions (Bat/Field)")
plt.show()
```

Player Statistics & Rankings


Top ten batsmen by total runs


```python
top_batsmen = deliveries.groupby("Batter")["Batter Runs"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_batsmen.index, y=top_batsmen.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Batsmen")
plt.ylabel("Total Runs")
plt.title("Top 10 Batsmen in IPL (Total Runs)")
plt.show()
```

Top 10 Bowlers by Wickets Taken



```python
top_bowlers = deliveries[deliveries["Wicket"].notna()].groupby("Bowler")["Wicket"].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_bowlers.index, y=top_bowlers.values, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("Bowlers")
plt.ylabel("Total Wickets")
plt.title("Top 10 Bowlers in IPL (Total Wickets)")
plt.show()
```

Venue & Toss Impact Study

Most winning venues

```python
venue_wins = matches["venue_name"].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=venue_wins.index, y=venue_wins.values, palette="pastel")
plt.xticks(rotation=90)
plt.xlabel("Stadium")
plt.ylabel("Matches Won")
plt.title("Top 10 Venues by Matches Won")
plt.show()
```
